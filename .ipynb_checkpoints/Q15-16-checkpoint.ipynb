```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df= np.loadtxt('hw4_nolabel_train.dat')
```

```python
E = []
V = []
K = [2,4,6,8,10]
for k in K:
    err = 0
    sq = 0
    for T in range(500):
        kmeans = KMeans(n_clusters=k, random_state=T).fit(df)
        err += kmeans.score(df)/100/500
        sq += ((kmeans.score(df)/100)**2)/500
    E.append(-1*err)
    V.append(sq - err**2)
```

```python
E
```

```python
K = [2,4,6,8,10]
```

```python
sns.set()
sns.lineplot(K,E)
print(E)
```

```python
sns.set()
sns.lineplot(K,V)
print(V)
```
